```python
def backtest(lst, holding_days):
    dct = {}
    for x in lst:
        starting = df[(df['TICKER'] == x[0]) & (df['date'] == x[-1])]
        const = 100 / (starting['OPENPRC'])
        var = (starting.index[0] + 1)
        for y in range(holding_days - 1):
            tmp = df.iloc[[var]]
            try:
                profit = round(float((float(df.iloc[[var - 1]]['OPENPRC']) - float(tmp['OPENPRC'])) * const), 2)
                if int(tmp['date']) in dct:
                    dct[int(tmp['date'])] += profit
                else:
                    dct[int(tmp['date'])] = profit
                #print(float(df.iloc[[var - 1]]['OPENPRC']), float(tmp['OPENPRC']), profit)
                if profit < -20:
                    print((tmp['TICKER']), profit, tmp['date'])
            except:
                pass
            var += 1
    return dct

fivedayholdingbacktest = backtest(backup, 5)
```
